# Unv 1-23

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re 
%matplotlib inline
```

```python
df = pd.read_csv('Unv_1_23.csv')
```

```python
df
```

## extract Unv number

```python
df['Unv.'] = df['title'].apply(lambda x : x.split(' ')[1])
```

```python
df['title']=df.apply(lambda x: x['title'].replace('Unv. ' + x['Unv.'],""), axis=1 )
```

```python
df
```

# exctract Biamonti number

```python
def Bia_number(text):
    match = re.search(r'\bBiamonti\b \d+',text)
    if match is not None:
#         print(match.group(0))
        df.loc[df['title']==text,['title']] = text.replace(match.group(0),'')
        return re.search('\d+',match.group(0)).group(0)
        
    else:
        return np.nan
```

```python
df['Biamonti'] = df['title'].apply(lambda x: Bia_number(x))
```

## exctract Hess number

```python
def Hess_number(text):
    match = re.search(r'\bHess\b \d+',text)
    if match is not None:
#         print(match.group(0))
        df.loc[df['title']==text,['title']] = text.replace(match.group(0),'')
        return re.search('\d+',match.group(0)).group(0)
        
    else:
        return np.nan
```

```python
df['Hess'] = df['title'].apply(lambda x: Hess_number(x))
```

## delete = in title

```python
df['title'] = df['title'].apply(lambda x : x.replace(' = ', '')[1:])
```

```python
df['title'].values
```

### delete - in key

```python
df['key'] = df['key'].apply(lambda x:x.replace('-',' '))
```

```python
df
```

## creat key_note

```python
df['key'] = df['key'].apply(lambda x : x[1:])
```

```python
df['key'].values
```

```python
def key_note(key):
    if key is not np.nan:
        spl = key.split(' ')
        print(spl)
        if len(spl)== 2:
            return spl[0]
        elif len(spl) == 3:
            return spl[0] + ' ' + spl[1]
    
    return np.nan
```

```python
df['key_note'] = df['key'].apply(key_note)
```

```python
df
```

## minor / major

```python
df['minor'] = df["key"].apply(lambda x : 1 if 'minor' in x else 0)
```

```python
df
```

## Create name

```python
def create_one_name(U, H, B):
    name = ''
    if H is not np.nan:
        name += 'Hess ' + H
    else:
        if U is not np.nan:
            name += 'Unv. ' + U
        
    if B is not np.nan:
        name += ', Biamonti ' + B
    
    return name
```

```python
df['name'] = df.apply(lambda x: create_one_name(x['Unv.'],x['Hess'],x['Biamonti']), axis=1)
```
