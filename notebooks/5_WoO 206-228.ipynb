```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
import re
```

```python
df = pd.read_csv('../csv/WoO_206_228.csv')
```

```python
df.head()
```

```python
df['WoO']=df['name'].apply(lambda x:x.split(':')[0].replace('WoO'," ").replace(' ',"").replace('No.',' No.'))
```

```python
df['WoO'].unique()
```

```python
all_names = df['name'].values
```

```python
all_names
```

## Create column with Hess number und create cleared Title for this composition

```python
for name in all_names:
    match = re.search('(\(Hess \d+\))', name)
    if match is not None:
        df.loc[df['name']==name,['Hess']] = re.search('\d+', match.group(0)).group(0)
        df.loc[df['name']==name,['Title']] = name.split(':')[1].replace(match.group(0),'')
    else:
        df.loc[df['name']==name,['Hess']] = np.nan
```

```python
df['Hess'].unique()
```

## Create column with Biamonti number und create cleared Title for this composition

```python
for name in all_names:
    match = re.search('(\(Biamonti \d+\))', name)
    
    if match is not None:
        df.loc[df['name']==name,['Biamonti']] = re.search('\d+', match.group(0)).group(0)
        df.loc[df['name']==name,['Title']] = name.split(':')[1].replace(match.group(0),'')
    else:
         df.loc[df['name']==name,['Biamonti']] = np.nan
    
```

```python
df['Biamonti'].unique()
```

```python
df.head(20)
```

## Clear title name for Gardi (i will not create Gardi column, cause in main db there is ot such catalogue name)

```python
for name in all_names:
    match = re.search('(\(Gardi \d+\))', name)
    if match is not None:
        df.loc[df['name']==name,['Title']] = name.split(':')[1].replace(match.group(0),'')
    
```

## Delete space from title

```python
df['Title'].unique()
```

```python
df['Title'] = df['Title'].apply(lambda x: x[1:-1])
```

## Create complex name

```python
df.drop(columns=['name'],inplace=True)
```

```python
df.info()
```

```python
def create_one_name(WoO, Hess, B):
    name = ''
    if WoO is not np.nan:
        name += 'WoO ' + WoO
    elif Hess is not np.nan:
            name += 'Hess ' + Hess

        
    if  type(B) != float and pd.isna(B)==False :
#         print(f"{B} {type(B)}" )
        name += ', Biamonti ' + str(B)
    
    return name
```

```python
df['name'] = df.apply(lambda x:create_one_name(x['WoO'],x['Hess'],x['Biamonti']), axis=1)
```

```python
df
```

## creat key_note

```python
df['key'] = df['key'].apply(lambda x : np.nan if x == ' ' else x)
```

```python
df['key'].values
```

delete space in the beggining

```python
df.loc[df['key'].isna()==False,['key']] = df.loc[df['key'].isna()==False,['key']]['key'].apply(lambda x : x[1:] if x[0]==' ' else x)
```

delete space in the end

```python
df.loc[df['key'].isna()==False,['key']] = df.loc[df['key'].isna()==False,['key']]['key'].apply(lambda x : x[:-1] if x[-1]==' ' else x)
```

```python
df.loc[df['key']=='in E flat major',['key']] = 'E flat major'
```

delete - in the key

```python
df.loc[df['key'].isna()==False,['key']] = df.loc[df['key'].isna()==False,['key']]['key'].apply(lambda x : x.replace('-',""))
```

```python
def key_note(key):
    if key is not np.nan:
        spl = key.split(' ')
#         print(spl)
        if len(spl)== 2:
            return spl[0]
        elif len(spl) == 3:
            return spl[0] + ' ' + spl[1]
    
    return np.nan
```

```python
df['key_note'] = df['key'].apply(key_note)
```

```python
df
```

## minor / major

```python
df['minor'] = df.loc[df['key'].isna()==False,['key']]["key"].apply(lambda x : 1 if 'minor' in x else 0)
```
