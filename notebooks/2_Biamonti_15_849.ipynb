## extract baimonti number

```python
def Bia_number(text):
    match = re.search(r'\bBia\b \d+\: ',text)
    if match is not None:
#         print(match.group(0))
        df.loc[df['title']==text,['title']] = text.replace(match.group(0),'')
        return re.search('\d+',match.group(0)).group(0)
        
    else:
        return np.nan
```

```python
df['Biamonti'] = df['title'].apply(lambda x: Bia_number(x))
```

## extract year

```python
def year(text):
    match = re.search('\(\d+\)',text)
    if match is not None:
        df.loc[df['title']==text,['title']] = text.replace(match.group(0),'')
        return re.search('\d+',match.group(0)).group(0)
    else:
        return np.nan
```

```python
df['year'] = df['title'].apply(lambda x:year(x))
```

## Extract key

```python
def key(text):
    match = re.search(r' (in|for) (C|D|E|F|G|A|B|H)',text)
    if match is not None:
        return match.group(0)[4:]
    else:
        return np.nan
```

```python
df['key'] = df['title'].apply(lambda x:key(x))
```

```python
def key_sign(text,key):
    match = re.search('(flat|sharp)',text)
    if match is not None:
        return key + ' ' + match.group(0)
    else:
        return key
```

```python
df['key'] = df.apply(lambda x: key_sign(x['title'], x['key']),axis=1)
```

```python
def key_major_minor(text,key):
    match = re.search(r'(major|minor)',text)
    if match is not None:
        return key + ' ' + match.group(0)
    else:
        return key
```

```python
df['key'] = df.apply(lambda x: key_major_minor(x['title'], x['key']),axis=1)
```

## extract key note

```python
df['key'].values
```

```python
df.loc[(df['key'].isna()==False),['key']] = df.loc[(df['key'].isna()==False),['key']]['key'].apply(lambda x : x[1:] if x[0] == " " else x )
```

```python
def key_note(key):
    if key is not np.nan:
        if key is not float:
            spl = key.split(' ')
    #         print(spl)
            if len(spl)== 2:
                return spl[0]
            elif len(spl) == 3:
                return spl[0] + ' ' + spl[1]
    
    return np.nan
```

```python
df['key_note']= df['key'].apply(key_note)
```

## extract minor

```python
df['minor'] = df.loc[(df['key'].isna()==False),['key']]['key'].apply(lambda x : 1 if 'minor' in x else 0)
```
